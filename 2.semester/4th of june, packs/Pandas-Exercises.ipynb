```python
import pandas as pd
filepath = "data/imdb-movies-dataset.csv"
df = pd.read_csv(filepath)  # read the data
```

```python
# Let's explore the dataframe using .columns and .describe()
print(df.columns,df.describe())
df.head()
```

```python
# Sort the dataframe by Rating, best one first!
df.sort_values(["Rating"],ascending=False)
```

```python
# Find out if Year and Rating are correlated

df["Year"].corr(df["Rating"])
```

```python
# Plot the year and rating as a scatter plot
import seaborn as sns
sns.set_theme()

sns.scatterplot(data=df,x=df["Year"],y=df["Rating"])
```

```python
sns.scatterplot(data=df,x=df["Director"],y=df["Rating"])
```

```python
sns.relplot(data=df,x=df['Duration (min)'],y=df["Rating"])
```

```python
sns.scatterplot(data=df,x=df['Duration (min)'],y=df["Rating"],)
```

```python
sns.displot(df,x=df["Director"],y=df["Metascore"])
```

```python
sns.lineplot(df,x=df["Rating"],y=df["Metascore"])
```

```python
sns.catplot(df, x=df["Review Count"], y=df["Metascore"], hue=df["Rating"])
```

```python
sns.jointplot(df,x=df["Genre"],y=df["Rating"])
```
